# *File*

نوشتن

```python
file = open("data.txt","w")
file .write ("Hello Python")

file.close()
```

خواندن

```python
file = open ("data.txt","r")
content = file.read()
print(content)
file.close()
```

اضافه کردن 

```python
file = open ("data.txt","a")
file .write ("\nNew line")
file.close()
```

روش حرفه ای تر

```python
with open ("data.txt","w") as file:
    file.write ("Hello Python")
```

کاربردی

```python
name = input ("نام خود را وارد کنید :")

with open ("names.txt","a") as file:
    file.write (name + "\n")
```
