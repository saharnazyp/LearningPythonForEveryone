```python
def play_game():
  secret = random.randint(1,100)
  tries = 0
  print ("=" *30)
  print ("بازی حدس عدد📲")
  print("=" * 30)
  print (" یه عدد بین 1 تا 100 انتخاب کردم ")
  print ("ُسعی کن حدس بزنی")
  print()

  while True:
    guess = int(input("حدس شما :"))
    tries += 1
    if guess < secret:
      print ("بیشتر!")
    elif guess > secret:
      print ("کمتر!")
    else:
      print (f"\n 🍺آفرین ! عدد {secret} بود!")
      print (f"در{tries} باز حدس بزن")
      if tries <= 5:
        print("عالی خیلی سریع بودی ")
      elif tries <= 10:
        print ("good")
      else:
        print("دفعه بعد بهتر میشی")
      break

play_game()
```
