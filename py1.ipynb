# *حلقه ها loops*

*for* برای تکرار روی لیست یا اعداد

```python
for i in range (5):
    print(i)
```

*while* تا وقتی شرط برقرار باشه

# Dictionaries دیکشنری ها 
یه دیکشنری (یا dict)    

توی پایتون ، داده ها رو به صورت کلید- مقدار 
(key - value)

نگه میدارد 

```python
person = {
    "name" :"sara",
    "age" : 32,
    "city" : "Tehran"
}

print(person ["name"])  #چاپ نام 
print (person["age"]) #چاپ سن 
```

اضافه  یا تغییر مقدار :

```python
person["job"]= "Data Scientist"  # اضافه کردن

person ["age"] = 33 #تغییر مقدار
```

حلقه روی دیکشنری:

```python
for key ,value in person.items ():
    print(key,":",value)
```
